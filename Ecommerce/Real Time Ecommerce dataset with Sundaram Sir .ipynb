```python
# recency = last date of transaction and today's date
import datetime as dt
current_date = dt.datetime(2017, 12, 20)
# frequency = total count of invoiceno on the basis of customerid
# monetary

RFMscore = dataset.groupby('CustomerID').agg({'Date' : lambda x : (current_date - x.max()).days,
                                            'InvoiceNo': lambda x : x.count(),
                                            'Price' : lambda x : x.sum()})

RFMscore.rename(columns = {"Date":"Recency", "InvoiceNo" : "Frequency", "Price" : "Monetory"}, inplace = True)
```

```python
# splitting the data using quantiles method
quantiles = RFMscore.quantile(q = [0.25, 0.5, 0.75])
quantiles = quantiles.to_dict()
quantiles
```

```python
def Rscores (x, p, d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4 
        
def FnMscores (x , p , d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.5]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1
```

```python
RFMscore['R'] = RFMscore['Recency'].apply(Rscores, args = ('Recency', quantiles))
RFMscore['F'] = RFMscore['Frequency'].apply(FnMscores, args = ('Frequency', quantiles))
RFMscore['M'] = RFMscore['Monetory'].apply(FnMscores, args = ('Monetory', quantiles))
```

```python
RFMscore
```

```python
RFMscore.reset_index()
```

```python
RFMscore['RFMgroup'] = RFMscore.R.map(str) + RFMscore.F.map(str) + RFMscore.M.map(str)
RFMscore['RFMgroup'] = RFMscore[['R','F','M']].sum(axis = 1)
RFMscore.reset_index()
```

```python
loyalty_level = ['Diamond','Gold','Platinum','Silver']
score_cut = pd.qcut(RFMscore.RFMgroup, q = 4, labels = loyalty_level)
RFMscore['Loyalty_level'] = score_cut.values
RFMscore = RFMscore.reset_index()
RFMscore
```

# Create Cluster Model

```python
RFMscore1 = RFMscore.copy()
```

```python
RFMscore1 = RFMscore.iloc[:, 1:4]
RFMscore1
```

```python
# feature engineering is required
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler_x = scaler.fit_transform(RFMscore1)
scaler_x
```

```python
# scaler_data = pd.DataFrame(scaler_x, index=RFMscore1.index, columns = RFMscore1.columns)
scaler_data = pd.DataFrame(scaler_x, index = RFMscore1.index, columns = RFMscore1.columns)
```

```python
scaler_data
```

```python
# Building Kmeans cluster
```

```python
# from sklearn.cluster import KMeans
# wcss = []

# for i in range(1,20):
#     kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 1000, random_state = 101)
#     kmeans.fit(scaler_data)
#     wcss.append(kmeans.inertia_)
    
    
from sklearn.cluster import KMeans
wcss = []

for i in range(1,20):
    kmeans = KMeans(n_clusters=i, init ='k-means++', max_iter = 1000, random_state=101)
    kmeans.fit(scaler_data)
    wcss.append(kmeans.inertia_)
```
